```python
G = nx.Graph()

G.add_edge('A', 'B', weight=4)
G.add_edge('B', 'D', weight=2)
G.add_edge('A', 'C', weight=3)
G.add_edge('C', 'D', weight=4)
# nx.shortest_path(G, 'A', 'D', weight='weight')

nx.draw(G, pos=nx.circular_layout(G), nodecolor='r', edge_color='b')
```

```python
G = nx.Graph()

e = [('a', 'b', 0.3), ('b', 'c', 0.9), ('a', 'c', 0.5), ('c', 'd', 1.2)]
G.add_weighted_edges_from(e)

nx.draw(G, pos=nx.circular_layout(G), nodecolor='lightblue', edge_color='b')
```

```python
G = nx.Graph()

e = [('1', '2', 1), 
     ('2', '3', 1), 
     ('3', '4', 1), 
     ('4', '2', 1)]

G.add_weighted_edges_from(e)

nx.draw(G, pos=nx.circular_layout(G), with_labels=True, nodecolor='b', edge_color='g')
```

```python
print('C = ', nx.clustering(G))
print('<C> = ', nx.average_clustering(G))
```

```python
G = nx.Graph()

e = [('01', '02', 1), 
     ('02', '03', 1), 
     ('03', '04', 1), 
     ('04', '05', 1), 
     ('05', '06', 1), 
     ('06', '07', 1), 
     ('07', '08', 1), 
     ('08', '09', 1), 
     ('09', '10', 1), 
     ('10', '11', 1), 
     ('11', '12', 1), 
     ('12', '01', 1)]

G.add_weighted_edges_from(e)

nx.draw(G, pos=nx.circular_layout(G), with_labels=True, nodecolor='b', edge_color='g')
```
